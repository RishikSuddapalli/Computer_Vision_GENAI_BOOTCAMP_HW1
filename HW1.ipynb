```python
import cv2
import numpy as np

def capture_average_brightness(video_path):
    # Accesing the video file
    video = cv2.VideoCapture(video_path)
    
    if not video.isOpened():
        raise ValueError("Error: Unable to open video file.")
    
    # Initialize a list to hold average brightness values for all frames
    avg_brightness_list = []
    
    frame_count = 0  # Counter for frames processed

    while True:
        ret, frame = video.read()
        if not ret:  # Break the loop if no more frames are available
            print("End of video.")
            break
        
        # Resizing the frame to fit a smartphone screen (720x1280)
        resized_frame = cv2.resize(frame, (720, 360))
        
        # Converting the resized frame to grayscale to calculate brightness
        gray_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2GRAY)
        
        # Calculating the average brightness of the frame
        avg_brightness = np.mean(gray_frame)
        avg_brightness_list.append(avg_brightness)
        
        # Display the resized frame
        cv2.imshow('Frame', resized_frame)
        print(f'Average Brightness: {avg_brightness:.2f}')
        
        # Wait for a short period to allow viewing of the frame
        key = cv2.waitKey(1)  # Display each frame for a short time (1 ms)
        
        frame_count += 1  # Increment the frame counter
        
        # Break the loop if 'q' is pressed
        if key == ord('q'):
            break
    
    # Release the video capture object
    video.release()
    cv2.destroyAllWindows()
    
    # Print the list of average brightness values
    print("Average Brightness List for all frames:")
    print(avg_brightness_list)
    print(f'Total frames processed: {len(avg_brightness_list)}')

# Example usage
capture_average_brightness("timelapse/Time_Lapse.mkv")


```

```python
import cv2
import numpy as np

def calculate_day_night_sunrise_sunset_percentage(video_path):
    # Define brightness thresholds
    brightness_day_threshold = 90    # Threshold for classifying frames as "day"
    brightness_sunrise_sunset_threshold = 30  # Threshold for sunrise/sunset
    
    # Open the video file
    video = cv2.VideoCapture(video_path)
    
    if not video.isOpened():
        raise ValueError("Error: Unable to open video file.")
    
    total_frames = 0
    day_frames = 0
    night_frames = 0
    sunrise_or_sunset_frames = 0
    
    
    while True:
        ret, frame = video.read()
        if not ret:  # Break the loop if no more frames are available
            break
        
        total_frames += 1
        
        # Convert the frame to grayscale to calculate brightness
        gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        
        # Calculate the average brightness of the frame
        avg_brightness = np.mean(gray_frame)
        
        # Classify the frame based on brightness
        if avg_brightness > brightness_day_threshold:
            day_frames += 1
        elif avg_brightness > brightness_sunrise_sunset_threshold:
            sunrise_or_sunset_frames += 1  # Frames that are bright but not fully day
        else:
            night_frames += 1
    
    # Release the video capture object
    video.release()
    
    # Calculate percentages
    day_percentage = (day_frames / total_frames) * 100 if total_frames > 0 else 0
    night_percentage = (night_frames / total_frames) * 100 if total_frames > 0 else 0
    sunrise_or_sunset_percentage = (sunrise_or_sunset_frames / total_frames) * 100 if total_frames > 0 else 0
    
    # Print the results in a dictionary format
    print({
        "day_percentage": day_percentage,
        "night_percentage": night_percentage,
        "sunrise_or_sunset_percentage": sunrise_or_sunset_percentage,
    })

# Example usage
calculate_day_night_sunrise_sunset_percentage("timelapse/Time_Lapse.mkv")
```
